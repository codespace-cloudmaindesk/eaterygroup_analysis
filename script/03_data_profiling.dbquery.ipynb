```sql
CREATE OR REPLACE TABLE eatery_group.silver.restaurant_profile AS
SELECT
  COUNT(*) AS total_rows,
  COUNT(DISTINCT restaurant_id) AS unique_restaurants,

  COUNT_IF(restaurant_id IS NULL) AS null_ids,
  COUNT_IF(restaurant_name IS NULL OR TRIM(restaurant_name) = '') AS null_names,
  COUNT_IF(city IS NULL OR TRIM(city) = '') AS null_cities,
  COUNT_IF(cuisines IS NULL OR TRIM(cuisines) = '') AS null_cuisines,
  COUNT_IF(aggregate_rating IS NULL) AS null_ratings,

  MIN(aggregate_rating) AS min_rating,
  MAX(aggregate_rating) AS max_rating,
  AVG(average_cost_for_two) AS avg_cost,

  COUNT_IF(latitude IS NULL OR latitude NOT BETWEEN -90 AND 90) AS invalid_lat,
  COUNT_IF(longitude IS NULL OR longitude NOT BETWEEN -180 AND 180) AS invalid_long,
  COUNT_IF(votes IS NULL OR votes < 0) AS invalid_votes

FROM eatery_group.bronze.restaurant_ds;

SELECT * FROM eatery_group.silver.restaurant_profile;
```
