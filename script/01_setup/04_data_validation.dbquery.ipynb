```sql
CREATE OR REPLACE TABLE eatery_group.silver.restaurant_validation AS
SELECT
  *,

  CASE WHEN restaurant_id IS NULL THEN 1 ELSE 0 END AS invalid_id,
  CASE WHEN restaurant_name IS NULL OR TRIM(restaurant_name) = '' THEN 1 ELSE 0 END AS invalid_name,
  CASE WHEN city IS NULL OR TRIM(city) = '' THEN 1 ELSE 0 END AS invalid_city,
  CASE WHEN cuisines IS NULL OR TRIM(cuisines) = '' THEN 1 ELSE 0 END AS invalid_cuisines,

  CASE WHEN latitude IS NULL OR latitude NOT BETWEEN -90 AND 90 THEN 1 ELSE 0 END AS invalid_latitude,
  CASE WHEN longitude IS NULL OR longitude NOT BETWEEN -180 AND 180 THEN 1 ELSE 0 END AS invalid_longitude,

  CASE WHEN average_cost_for_two < 0 THEN 1 ELSE 0 END AS invalid_cost,
  CASE WHEN votes IS NULL OR votes < 0 THEN 1 ELSE 0 END AS invalid_votes,
  CASE WHEN aggregate_rating IS NULL OR aggregate_rating NOT BETWEEN 0 AND 5 THEN 1 ELSE 0 END AS invalid_rating

FROM eatery_group.bronze.restaurant_ds;

SELECT * FROM eatery_group.silver.restaurant_validation;
```
